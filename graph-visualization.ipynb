```python
import pandas as pd

from graphrag.query.indexer_adapters import (
    read_indexer_entities,
    read_indexer_relationships,
)
```

```python
INPUT_DIR = "./ex1/output"

entity_df = pd.read_parquet(f"{INPUT_DIR}/entities.parquet")
entity_df
```

```python
relationship_df = pd.read_parquet(f"{INPUT_DIR}/relationships.parquet")
relationship_df.head()
```

```python
community_df = pd.read_parquet(f"{INPUT_DIR}/communities.parquet")
community_df.head()
```

```python
community_reports_df = pd.read_parquet(f"{INPUT_DIR}/community_reports.parquet")
community_reports_df.head()
```

```python
document_df = pd.read_parquet(f"{INPUT_DIR}/documents.parquet")
document_df.head()
```

```python
text_units_df = pd.read_parquet(f"{INPUT_DIR}/text_units.parquet")
text_units_df.head()
```

```python
entities = read_indexer_entities(entity_df, community_df , None)

entities_dict = []
for e in entities:
    entity = {
        "id": e.title,
        "properties":{
            "id": e.id,
            "title": e.title,
            "type": e.type,
            "description": e.description,
            "community_ids": e.community_ids,
            "text_unit_ids": e.text_unit_ids,
            "rank": e.rank
            }
        }
    entities_dict.append(entity)

# entities_dict
```

```python
relationships = read_indexer_relationships(relationship_df)
relationships_dict = []
for r in relationships:
    relationship = {
        "start": r.source,
        "end": r.target,
        "properties": {
            "id": r.id,
            "source": r.source,
            "target": r.target,
            "description": r.description,
            "text_unit_ids": r.text_unit_ids,
            "rank": r.rank,
            "weight": r.weight
        }
    }
    relationships_dict.append(relationship)
# relationships_dict
```

```python
from yfiles_jupyter_graphs import GraphWidget

w = GraphWidget()
w.directed = True
w.nodes = entities_dict
w.edges = relationships_dict
w.node_label_mapping = "title"
```

```python
# map community to a color
def community_to_color(community):
    """Map a community to a color."""
    colors = [
        "crimson",
        "darkorange",
        "indigo",
        "cornflowerblue",
        "cyan",
        "teal",
        "green",
    ]
    return (
        colors[int(community) % len(colors)] if community is not None else "lightgray"
    )


def edge_to_source_community(edge):
    """Get the community of the source node of an edge."""
    source_node = next(
        (entry for entry in w.nodes if entry["properties"]["title"] == edge["start"]),
        None,
    )
    source_node_community = source_node["properties"]["community_ids"]
    return source_node_community if source_node_community is not None else None


w.node_color_mapping = lambda node: community_to_color(node["properties"]["community_ids"][0])
#w.edge_color_mapping = lambda edge: community_to_color(edge_to_source_community(edge))
# map size data to a reasonable factor
w.node_scale_factor_mapping = lambda node: 0.5 + node["properties"]["rank"] * 1.5 / 20
# use weight for edge thickness
w.edge_thickness_factor_mapping = "weight"
w.circular_layout()
```
